```python
geolocator = Nominatim(user_agent="athena")

city ="Chicago"
state ="Illinois"

loc = geolocator.geocode(city+','+ state)

print("latitude is :" ,loc.latitude,"\nlongtitude is:" ,loc.longitude)
```

```python
cities_df_path = '/Users/evanmcneal/Desktop/Projects/athena/data_igsn/data/cities.csv'
```

```python
cities = pd.read_csv(cities_df_path)
```

```python
cities
```

```python
for index, row in cities.iterrows():
    
    if pd.isna(row["latitude"]):
        city = row["City"]
        state = row["State"]

        loc = geolocator.geocode(city + ',' + state)

        cities.loc[index, "latitude"] = loc.latitude
        cities.loc[index, "longitude"] = loc.longitude
```
